```python
import pygame
import time

pygame.init()

window_width = 800
window_height = 600
fon = 'apple.png'

window = pygame.display.set_mode((window_width, window_height))

speed = 0  # текущая скорость перемещения
sdvig_fon = 0  # сдвиг фона

# загрузка и подготовка фона
try:
    fon_img = pygame.image.load(fon)
    back_fon = pygame.transform.scale(fon_img, (window_width, window_height))
except:
    print("Ошибка фона")
    pygame.quit()
    exit()

running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        
        # обработка нажатий клавиш
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                speed = 3  # движение вправо 
            elif event.key == pygame.K_RIGHT:
                speed = -3  # движение влево 
        
        if event.type == pygame.KEYUP:
            if event.key == pygame.K_LEFT or event.key == pygame.K_RIGHT:
                speed = 0

    # обновление сдвига фона
    sdvig_fon += speed
    
    # обеспечение бесконечного скроллинга
    if sdvig_fon >= window_width:
        sdvig_fon = 0
    elif sdvig_fon <= -window_width:
        sdvig_fon = 0
    
    # отрисовка фона
    window.fill((0, 0, 0))  # очистка экрана
    window.blit(back_fon, (sdvig_fon, 0))
    
    #для бесшовного скроллинга рисуем второй экземпляр фона
    if sdvig_fon > 0:
        window.blit(back_fon, (sdvig_fon - window_width, 0))
    else:
        window.blit(back_fon, (sdvig_fon + window_width, 0))
    
    pygame.display.update()
    time.sleep(0.02)

pygame.quit()
```

```python
import pygame
import sys

pygame.init()

window_width = 800
window_height = 600
window = pygame.display.set_mode((window_width, window_height))


# класс игрока
class Player(pygame.sprite.Sprite):
    def __init__(self, filename, hero_x=100, hero_y=250, x_speed=0, y_speed=0):
        pygame.sprite.Sprite.__init__(self)
        self.image = pygame.image.load(filename)  # загрузка изображения
        self.rect = self.image.get_rect()
        self.hero_x = hero_x
        self.hero_y = hero_y
        self.rect.x = hero_x  # начальная позиция X
        self.rect.y = hero_y  # начальная позиция Y
        self.x_speed = x_speed  # скорость по X
        self.y_speed = y_speed  # скорость по Y
    
    def update(self):
        # перемещает персонажа с текущей скоростью
        self.rect.x += self.x_speed
        self.rect.y += self.y_speed

# класс фона
class Background:
    def __init__(self, filename, width, height):
        self.image = pygame.image.load(filename)
        self.image = pygame.transform.scale(self.image, (width, height))
        self.rect = self.image.get_rect()
        self.speed = 0
        self.x = 0
    
    def update(self):
        # обновляет позицию фона
        self.x += self.speed
        if self.x >= window_width:
            self.x = 0
        elif self.x <= -window_width:
            self.x = 0
    
    def draw(self, surface):
        # отрисовывает фон с учетом смещения
        surface.blit(self.image, (self.x, 0))
        # для бесконечного фона
        if self.x > 0:
            surface.blit(self.image, (self.x - window_width, 0))
        else:
            surface.blit(self.image, (self.x + window_width, 0))

try:
    # создаем игрока
    player = Player('apple.png')
    all_sprites = pygame.sprite.Group(player)
    
    # Создаем фон
    background = Background('fon.png', window_width, window_height)
    
except pygame.error as e:
    print(f"Ошибка загрузки изображений: {e}")
    pygame.quit()
    sys.exit()

# игровой цикл
clock = pygame.time.Clock()
running = True

while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        
        # обработка нажатий клавиш
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                player.x_speed = -5
            elif event.key == pygame.K_RIGHT:
                player.x_speed = 5
            elif event.key == pygame.K_UP:
                player.y_speed = -5
            elif event.key == pygame.K_DOWN:
                player.y_speed = 5
        
        if event.type == pygame.KEYUP:
            if event.key in (pygame.K_LEFT, pygame.K_RIGHT):
                player.x_speed = 0
            if event.key in (pygame.K_UP, pygame.K_DOWN):
                player.y_speed = 0
    
    # проверка границ для движения фона
    if player.rect.right >= window_width and player.x_speed > 0:
        background.speed = -5  # фон движется влево
        player.x_speed = 0
    elif player.rect.left <= 0 and player.x_speed < 0:
        background.speed = 5  # фон движется вправо
        player.x_speed = 0
    else:
        background.speed = 0
    
    # обновление объектов
    all_sprites.update()
    background.update()
    
    # отрисовка
    window.fill((0, 0, 0))  # очистка экрана
    background.draw(window)
    all_sprites.draw(window)
    
    pygame.display.flip()
    clock.tick(60)

pygame.quit()
```

```python
import pygame
import sys
import random

pygame.init()


window_width = 800
window_height = 600
window = pygame.display.set_mode((window_width, window_height))

BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
RED = (255, 0, 0)

# класс игрока
class Player(pygame.sprite.Sprite):
    def __init__(self, filename, x=100, y=250):
        pygame.sprite.Sprite.__init__(self)
        self.image = pygame.image.load(filename)
        self.rect = self.image.get_rect()
        self.rect.x = x
        self.rect.y = y
        self.speed_x = 0
        self.speed_y = 0
    
    def update(self):
        self.rect.x += self.speed_x
        self.rect.y += self.speed_y
        
        # ограничение движения по границам экрана
        if self.rect.right > window_width:
            self.rect.right = window_width
        if self.rect.left < 0:
            self.rect.left = 0
        if self.rect.top < 0:
            self.rect.top = 0
        if self.rect.bottom > window_height:
            self.rect.bottom = window_height

# класс врага
class Enemy(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.image = pygame.Surface((30, 30))
        self.image.fill(RED)
        self.rect = self.image.get_rect()
        self.rect.x = random.randrange(window_width - self.rect.width)
        self.rect.y = random.randrange(window_height - self.rect.height)
        self.speed_x = random.randrange(-3, 3)
        self.speed_y = random.randrange(-3, 3)
    
    def update(self):
        self.rect.x += self.speed_x
        self.rect.y += self.speed_y
        
        # Отскок от границ экрана
        if self.rect.right > window_width or self.rect.left < 0:
            self.speed_x *= -1
        if self.rect.bottom > window_height or self.rect.top < 0:
            self.speed_y *= -1

# класс стрелы
class Arrow(pygame.sprite.Sprite):
    def __init__(self, x, y):
        pygame.sprite.Sprite.__init__(self)
        self.image = pygame.Surface((10, 5))
        self.image.fill(WHITE)
        self.rect = self.image.get_rect()
        self.rect.left = x
        self.rect.centery = y
        self.speed = 10
    
    def update(self):
        self.rect.x += self.speed
        # Удаление стрелы, если она вышла за экран
        if self.rect.left > window_width:
            self.kill()

# создание спрайтов
all_sprites = pygame.sprite.Group()
enemies = pygame.sprite.Group()
arrows = pygame.sprite.Group()

# главный герой
player = Player('apple.png')
all_sprites.add(player)

# враги
for i in range(5):
    enemy = Enemy()
    all_sprites.add(enemy)
    enemies.add(enemy)

# Игровой цикл
clock = pygame.time.Clock()
running = True

while running:
    # обработка событий
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        
        # управление игроком
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                player.speed_x = -5
            elif event.key == pygame.K_RIGHT:
                player.speed_x = 5
            elif event.key == pygame.K_UP:
                player.speed_y = -5
            elif event.key == pygame.K_DOWN:
                player.speed_y = 5
            elif event.key == pygame.K_SPACE:
                # Создание стрелы при нажатии пробела
                arrow = Arrow(player.rect.right, player.rect.centery)
                all_sprites.add(arrow)
                arrows.add(arrow)
        
        if event.type == pygame.KEYUP:
            if event.key in (pygame.K_LEFT, pygame.K_RIGHT):
                player.speed_x = 0
            if event.key in (pygame.K_UP, pygame.K_DOWN):
                player.speed_y = 0
    
    # обновление спрайтов
    all_sprites.update()
    
    # проверка столкновения стрел с врагами
    hits = pygame.sprite.groupcollide(enemies, arrows, True, True)
    for hit in hits:
        # создаем нового врага вместо уничтоженного
        enemy = Enemy()
        all_sprites.add(enemy)
        enemies.add(enemy)
    
    # отрисовка
    window.fill(BLACK)
    all_sprites.draw(window)
    pygame.display.flip()
    clock.tick(60)

pygame.quit()
sys.exit()
```
